### Mengenal OO style (Object Oriented Style)

```python
x = np.linspace (0,2,100)
x 
```

```python

# Buat figure dan axes
fig, ax = plt.subplots()

# Plot tiga garis
ax.plot(x, x, label="Linear")
ax.plot(x, x**2, label="Quadratic")
ax.plot(x, x**3, label="Cubic")

# Label sumbu
ax.set_xlabel("X Label")  # untuk sumbu X
ax.set_ylabel("Y Label")  # untuk sumbu Y

# Judul & legenda
ax.set_title("Simple Plot")
ax.legend()

# Tampilkan grafik
plt.show()
```

### Mengenal Pyplot style


```python
x = np.linspace(0,2,100)
x
```
