
## 02 Pemilihan Baris( Row Selection ) Pada Data Frame

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1,10, size =(n_rows, n_cols)),
                  columns = cols)
df
```

```python
#selection dengan operator logika |
df[(df['A']==1)|(df['A']==3)]
```

```python
#selection dengan fungsi isin()

df[df['A'].isin([1,3])]
```

```python
#mengenal operator negasi -
df[-df['A'].isin([1,3])]
```

### 04 Pemilihan Kolom pada Pandas DataFrame Berdasarkan tipe data

```python
import pandas as pd
import numpy as np

# Buat DataFrame
n_rows = 5
n_cols = 2
cols = ['bil_pecahan', 'bil_bulat']
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

# Ubah kolom bil_pecahan jadi float
df['bil_pecahan'] = df['bil_pecahan'].astype(float)

# Buat index berupa waktu (tiap 1 jam)
df.index = pd.date_range(start='2025-10-14', periods=n_rows, freq='H')

# Reset index jadi kolom biasa (opsional)
df = df.reset_index()

# Tambahkan kolom baru bernama 'teks'
df['teks'] = list('ABCDE')

# Tampilkan DataFrame
df
```

##### Memilih kolom bertipe data numerik

```python
df.select_dtypes(include='number')
```

```python
df.select_dtypes(include='float')
```

```python
df.select_dtypes(include='int')
```

##### Memilih kolom bertipe data string atau object

```python
df.select_dtypes(include='datetime')
```

##### Memilih kolom bertipe dengan kombinasi tipe data

```python
df.select_dtypes(include=['number','object'])
```
