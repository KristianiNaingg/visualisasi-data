### 03 Multiple Subplots

```python
%matplotlib inline
```

```python

import matplotlib
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

##### Simple Line Plot

```python
x = np.arange(.0,2.0, 0.01)
x
```

```python
s = np.sin(2*np.pi* x)
s
```

```python
fig, ax = plt.subplots()

# Membuat plot
ax.plot(x, s)

# Menetapkan label dan judul
ax.set(xlabel='nilai x', ylabel='nilai sin(x)', title='Visualisasi nilai sin')

# Menampilkan grid
ax.grid()

# Menampilkan plot
plt.show()
```

##### 03 Matplotlib- Multiple Subplot

```python
# Membuat array x
x1 = np.linspace(0.0, 5.0, 100) # 100 titik dari 0 hingga 5
x2 = np.linspace(0.0, 2.0, 100) # 100 titik dari 0 hingga 2

# Menghitung nilai kosinus
y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

##### Multiple Subplots dengan OO Style

```python
# Membuat figure (fig) dan dua axes/subplot (ax1, ax2) dalam konfigurasi 2 baris, 1 kolom
fig, (ax1, ax2) = plt.subplots(2, 1)

# Menambahkan judul utama untuk figure
fig.suptitle('Multiple Subplots')

# --- Konfigurasi Subplot 1 (ax1) ---
# ax1.plot(data x1, data y1, format: 'r-o' = merah, garis solid, marker lingkaran)
ax1.plot(x1, y1, 'ro-') 
ax1.set_ylabel('Nilai cos(x1)')

# --- Konfigurasi Subplot 2 (ax2) ---
# ax2.plot(data x2, data y2, format: 'g--' = hijau, garis putus-putus)
ax2.plot(x2, y2, 'g--')
ax2.set_ylabel('Nilai cos(x2)')
ax2.set_xlabel('Nilai x') # Label x hanya ditempatkan di subplot paling bawah

# Menampilkan figure
plt.show()
```

##### Multiple Subplots dengan pyplot Style

```python
# Multiple Subplots dengan Pyplot Style (image_4900c1.png)

# Subplot 1 (Baris 2, Kolom 1, Posisi 1)
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'ro-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai $\cos(x\_1)$')

# Subplot 2 (Baris 2, Kolom 1, Posisi 2)
plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'g--')
plt.ylabel('Nilai $\cos(x\_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```

### 04 MatplotLib

```python
# Mu (mean) = 100, Sigma (standard deviation) = 15
mu, sigma = 100, 15 

# Membuat 10000 nilai dengan distribusi normal
x = mu + sigma * np.random.randn(10000)
x

```

```python
x.shape
```

##### Histogram dengan pyplot Style

```python
plt.hist(x, 
         bins=50, 
         facecolor='g', 
         alpha=0.75)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

# Menambahkan teks anotasi (menggunakan format LaTeX untuk mu dan sigma)
plt.text(45, 500, r'$\mu=100,\ \sigma=15$') # Catatan: nilai posisi (45, 500) mungkin perlu disesuaikan dengan data acak
plt.grid()

plt.show()
```

##### Histogram dengan OO Style

```python

fig, ax = plt.subplots()

ax.hist(x, 
        bins=50, 
        facecolor='r', 
        alpha=0.75)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

# Menambahkan teks anotasi (menggunakan format LaTeX untuk mu dan sigma)
ax.text(45, 500, r'$\mu=100,\ \sigma=15$')
ax.grid()

plt.show()
```
