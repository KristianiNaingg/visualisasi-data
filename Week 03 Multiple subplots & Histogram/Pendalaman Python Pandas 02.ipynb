### 10 Mengganti Nama Kolom Data Frame menurut Pola

```python
df = pd.read_csv('./data/titanicfull.csv')
df.columns = ['Pclass','Survival status','full Name','Sex','Age','Sib SP','Parch','Ticket','Fare','Cabin','Embarked']

df_backup = df.copy(deep=True)
df.head()
```

##### Using lowercase for column names and replacing spaces with _

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
df.head()
```

##### Memangkasa kelebihan spasi pada nama kolom

```python
df = df_backup.copy(deep=True)

df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
df.head()
```

### 11 Seleksi Baris & Kolom Data Frame dengan loc

```python
n_rows = 10
n_cols = 5

cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                 columns=cols)
df
```

##### Seleksi kolom dan baris menggunakan loc

```python
df.loc[[0, 3, 4], ['B', 'E']]
```

##### Seleksi Baris dengan Kondisi

```python
df.loc[df['B'] > 10 ,['B', 'D', 'E']]
```

##### Slicing Data Frame dengan loc

```python
df.loc[0:4, 'B':'D']
```

### 12 Membentuk Kolom datetime dari beberapa Kolom Data Frame

```python
data = {'day':[1, 2, 10, 25, 12],
        'month':[1, 2, 4, 5, 6],
        'year':[2000, 2001, 2010, 2015, 2020]}
df = pd.DataFrame(data)
df
```

#### Membentuk Kolom bertipe datetime

```python
df['penanggalan'] = pd.to_datetime(df[['day', 'month', 'year']])
df
```
