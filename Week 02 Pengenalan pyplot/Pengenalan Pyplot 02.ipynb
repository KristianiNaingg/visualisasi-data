```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

##### Membuat Plotting Sederhana

```python
plt.plot([2, 5, 7, 11])

plt.ylabel('sumbu y')
plt.show()
```

```python
plt.plot([2, 3, 4, 7], [5, 4, 2, 16])
plt.show
```

##### Pengaturan Format pada plot

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

##### Multi Plot pada satu pemanggilan plot()


```python
t = np.arange(0., 5., 0.2)
t
```

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
## atau kita bisa juga multi plot dengan cara ini 
plt.plot(t, t, 'r--',
         t, t**2, 'bs',
         t, t**3, 'g^')
plt.show()
```

##### Plot dengan keywords


```python

data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
data
```

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

##### Plotting untuk tipe data categorical 

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
```

```python
plt.plot(names, values)
plt.show()
```

```python
plt.scatter(names, values)
plt.show()
```

```python
plt.bar(names, values)
plt.show()
```
